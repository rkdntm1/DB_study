```python
import pandas as pd

data = {"name":['a1','a2',np.nan,'a4']
       ,"age":[10,np.nan,np.nan,20]
       ,"addr":['seoul',np.nan,np.nan,'seoul']
       ,"d1":[1,2,np.nan,4]}
data =pd.DataFrame(data, columns=['name','age','addr','d1'], index=['t1','t2','t3','t4'])
data
```

```python
# 결측치가 몇개 있는지 확인 
data.isnull().sum()
```

```python
#dropna : 결측치값 존재시 해당 로우 삭제
data.dropna()
```

```python
# dropna 조건붙이기
data.dropna(how = 'all') # how = 'all' : 모든 정보들이 nan 값이면 삭제  
```

```python
data.dropna(how='all', inplace = True) # inplace 를 true로 해서 자체수정
```

```python
# nan값 수정하기 fillna 
data['age'].fillna(data['age'].mean(), inplace= True)
```

```python
data['addr'].fillna('seoul', inplace= True)
```

```python
data
```

```python
# any() : 결측치가 있는지 각 칼럼별 유무 확인
data.isnull().any()
```

```python
# 전체적인 결측치가 있는지 유무 확인 
data.isnull().any().any()
```

```python

def histogram_intersection(a, b):
    v = np.minimum(a, b).sum().round(decimals=1)
  
    
animal = pd.DataFrame([(.2, .3), (.0, .6), (.6, .0), (.2, .1)],
                  columns=['dogs', 'cats'])
animal.corr(method=histogram_intersection)
```

```python
import pandas as pd
import numpy as np

d1=pd.DataFrame([['hong',20],['kim',13]],columns=['name','age'], index=['a1','a2'])
print(d1)
d2=pd.DataFrame([['hong',100],['kim',20],['park',np.nan]],columns=['name','jumsu'], index=['a1','a2','a3'])

print(d2)
pd.merge(d1, d2,how='inner', on='name')
```

```python
user = {'hakno':[1,2,3,4,5,6]
     ,'name':['hong','kim','lee','park','han','kang']
     }
product_user ={'hakno':[1,2,3,2,2]
               ,'productname':['p1','p2','p3','p4','p2']
               ,'ordersu':[10,20,10,1,2]
              }
d1 = pd.DataFrame(user,columns=['hakno','name'])
d2 = pd.DataFrame(product_user, columns=['hakno','productname','ordersu'])

result = pd.merge(d1, d2, on = 'hakno', how ='inner')
result
```

```python
#pivot 데이터의 정보를 따로 뽑아서 옆에다 집어넣음
from pandas import DataFrame

data=[['20201110','A',10],
      ['20201112','B',15],
      ['20201112','A',20],
      ['20201201','A',20],
      ['20201203','B',50]
     ]
df = pd.DataFrame(data,columns=['date','productid','data'])
print(df)
print()
pivot1=df.pivot(index = 'date', columns='productid', values='data')
print(pivot1)
```

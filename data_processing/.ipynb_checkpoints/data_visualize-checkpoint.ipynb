```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1,5, 11)
y=x**2

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x and y ( x**2) graph')
plt.show()
```

```python
#한글 깨짐 폰트 수정 #1 
import sys
import numpy as np
import matplotlib.pyplot as plt

print(sys.platform)
fontname = np.nan
if sys.platform in ['win32', 'win64']:
    fontname ='malgun gothic'
plt.rc('font', family=fontname)
    
x = np.linspace(1,5, 11)
y=x**2

plt.plot(x, y)
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('x and y ( x**2) 그래프')
plt.show()
```

```python
#한글 깨짐 폰트 수정 #2 
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm


path = "c:/windows/fonts/gulim.ttc"
font_name = fm.FontProperties(fname =path,size=10).get_name()

plt.rc('font', family=font_name)

x = np.linspace(1,5, 11)
y=x**2

plt.plot(x, y)
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('x and y ( x**2) 그래프')
plt.show()
```

```python
#다수 그래프 그리기

import numpy as np
import matplotlib.pyplot as plt

x1 = np.arange(0,5,0.5)
x2 = np.arange(0,5,0.5)

plt.plot(x1, x1, 'b--', x2, x2**2,'rd',x1,x1**3,'cv-')
plt.xlabel('x-data')
plt.ylabel('y-data')
plt.title('graph')
plt.show()
```

```python
# subplots
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()

plt.subplot()
plt.subplots(2,1,1)


x = np.arange(1,5)
y = x**2

plt.plot(x,y)

plt.subplots(2,1,2)

x = np.arange(1,5)
y = x**3

plt.plot(x,y)
plt.show()

```


# tsv 파일을 dict타입의 자료로 저장

```python
from collections import defaultdict

data = defaultdict(list)
input_file_name = './data/seoul_wifi.tsv'
with open(input_file_name, 'r', encoding='utf8') as input_f:
    field_name = None
    for i,line in enumerate(input_f): #enumerate : 인덱스를 추가할때 사용
        #print('i: {0}, line: {1}'.format(i, line))
        like = line.strip()     
        #print(like)        
        lite = like.split('\t')
        #print(lite)
        
        if i ==0:
            field_name = lite
            continue
            
        if i==10:
            break
    
        for element, b in zip(lite, field_name): #filed_name :필드 이름, element :각요소
            data[b].append(element)
data
```

```python
from collections import defaultdict

data = defaultdict(list)
input_file = './data/member.csv'
with open(input_file, 'r', encoding = 'utf8') as t:
    lines = csv.reader(t)
    for i, line in enumerate(lines):
        #print(i, line)
        #line =line.split(',')        
        #print(line)
        if i == 0 :
            field_name = line
            continue
        for a, b in zip(field_name, line):
            data[a].append(b)
data
```

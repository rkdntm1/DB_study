```python
from bs4 import BeautifulSoup
import requests  

USER_AGENT="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36"
HEADERS={"User-Agent":USER_AGENT}
number_data=[]
    
def get_text(url):
    data=requests.get(url)
    page_content=data.text
    return page_content

def get_data(data):
    soup=BeautifulSoup(data,'lxml')
    _box=soup.find('table',class_="type_1")
    ctg_data=_box.find_all('td',class_="ctg") # 이 부분의 텍스트 파일이 종목명
   
 #sibling 으로 동일선 상의 다음 'td' 값을가져옴
    number_data=[index.find_next_sibling('td')  for index in ctg_data] 
    return ctg_data, number_data


if __name__=='__main__':
    for i in range(1, 20):
        url="https://finance.naver.com/sise/entryJongmok.nhn?&page={0}".format(i)
        page_content=get_text(url)
        ctg_data, number_data=get_data(page_content)
    
        for i, j in zip(ctg_data, number_data):
             print('종목별 :{0} \t 현재가:{1}'.format(i.text.strip(),j.text.strip()))
    
    
```

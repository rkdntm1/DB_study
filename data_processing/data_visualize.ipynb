```python
import numpy as anp
import matplotlib.pyplot as plt

x1= np.arange(1,5)
y1= np.exp(x1)

x2= np.arange(1,10,2)
y2= x2**3

plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.title('x-y(exp(x)) graph ')
plt.xlabel('x')
plt.ylabel('y')
plt.text(np.mean(x1), np.mean(y1), 'mean')
plt.grid(True) #grid :격자

plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title('x-y(x**3) graph ')
plt.xlabel('x...')
plt.ylabel('y...')
plt.grid(False)
```

```python
# 랜덤 데이터 
import numpy as np
np.random.seed(19680801)
import matplotlib.pyplot as plt


fig, ax = plt.subplots()
for color in ['tab:orange','tab:blue', 'tab:green']:
    n = 50
    x, y = np.random.rand(2, n)
    scale = 200.0 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color, 
               alpha=0.3, edgecolors='none')

ax.legend(loc="upper right")
ax.grid(True)

plt.show()
```

```python
bins = 20
import scipy as sp
plt.hist(x=tips.tip, rwidth=0.9, bins=bins, density=True)

mu, sigma = tips.tip.mean(), tips.tip.std()


tip_min = tips.tip.min()
tip_max = tips.tip.max()

x=np.linspace(tip_min, tip_max, bins)
y=sp.stats.norm(mu, sigma).pdf(x)
plt.plot(x,y)
```

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

mu, sigma = 100, 15  # mu:평균  sigma: 표준편차
x = mu + sigma*np.random.randn(10000)
# the histogram of the data

plt.hist(x, 50, density=True, color='green', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```

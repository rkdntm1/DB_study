##  1. 단일 회귀분석(simple linear regression model)



$Y_i = \beta_0 + \beta_1X_i+\epsilon_i$ 

- $Y$ : response/output/target(종속변수) 
- $X$ : feature/input/predictor(독립변수)
- $\beta_0$ is the intercept
- $\beta_1$ 회귀계수(regression coefficient)  
- $\epsilon$ 오차항, 노이즈  
  회귀계수-regression coefficient


### $e_i$= $y_i - \hat y_i$  : 잔차 (residual)   


#### 오류합에 절대값을 취하여 더하는 방식(Mean Absolute Error)

#### 오류값의 제곱을 더해서 구하는 방식(RSS :Residual Sum of Square)
  => 미분이 용이하므로 많이 사용
  
  $Error^2$=RSS


### 최소제곱- 실제 Y값과 예측 된 Y값의 차이
### => Sum of Squared Error :SSE 

#### 예측과 훈련 세트에 있는 타깃 y사이의 평균제곱 오차   E (Y- $ \hat Y)^2 $   
#### ( MSE :  mean sequred error)를 최소화 하도록 함
  
  ###   $ \frac{1}{n} \sum_{i=0}^{n}(Y_i- \hat Y_i)^2 $  :  평균제곱 오차( MSE :  mean sequre error)
  
 #### RMSE(root mean sequred error) : MSE 에 루트  씌운것 
#### MSLE(mean sequared log error) :  MSE에 로그 취한 것
#### RMSLE(root mean sequed log error) 



#### SST(Total sum of sequares) : $  \sum_\ (y_i - \bar y_i)^2 $      

#### SSE(Sum of squareds due to residual errors) : $ \sum_ \ (y_i- \hat y_i)^2 $  : 잔차에 의한 제곱합

#### SSR(sum of squares due to regression) : $ \sum (\hat y_i - \bar y_i) ^2  $  : 회귀에 의한 제곱합

### SST(총변동)= SSE(회귀선에서 설명이 안되는 변동) + SSR (회귀선에서 설명되는 변동)



### 결정계수 : $ r^2 $   

$ r^2 $ = $ \frac{ SSE }{ SST } $   =  1-  $ \frac{SSR }{SST} $ (   범위 0<= $ R^2$ <=1)    

학습한 회귀모델이 얼마나 데이터를 잘 표현하는지에 대한  정도를  
나타내는 통계적 척도  
=> X로 Y의 분산도를 얼마나 잘 설명할 수 있는가를 나타냄

 $R^2$이 1에 가까울 수록 x와 y사이에 높은 상관관계를 갖고 있다.  
   ( 모델이 데이터를 잘 표현하고 있다)     
   
  0에 가까운 값을 가지는 회귀선은 모델이 데이터를 표현하지 못한다는 것을  
  의미한다.

## 2. 다중 회귀분석( Multiple Linear Regression)

 ###  $ y_i = \beta_0 + \beta_1 x_{i1} +\beta_2 x_{i2}+ ... +\beta_p x_{ip} + \epsilon  $  
   * i = 1,2,...n개의 샘플, p=1,2,...p개의 feature




#### train 데이터(X, Y ) 를 이용해  parameter $ \beta_0, .... \beta_p $ 의 추정치 
 $ \hat\beta_0, \hat \beta_1... \hat \beta_p $를 구함

####  $ y_i$ 를 response라 하며,  $ x_{ij}$ (j=1,2....p)를 feature라 함

$ y = w_0x_0 + w_1x_1 + w_2x_2 + .... w_nx_n + b$   
$ w : 계수(가중치) $    
$ b : 절편(편향) $

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x_train, x_test, y_train, y_test = train_test_split(df, target, test_size=0.2, random_state=2020)
#x_train, x_test, y_train, y_test = train_test_split(df, target, test_size=0.2, random_state=0)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
print('train set {0}'.format(lr.score(x_train, y_train)))
print('predict set {0}'.format(lr.score(x_test, y_test)))
```

```python
mse=mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)

print("mse:{0}, rmse :{1}".format(mse, rmse))
```

```python
r2=r2_score(y_test, y_pred)
print('variance score: {0}'.format(r2))
```

```python
print('절편값 {0}'.format(lr.intercept_))
print('회귀계수 :{0}'.format(np.round(lr.coef_,1)))
```

회귀분석에서 비용함수는 MSE를 최소화하는 방향으로 가는데    
데이터의 차원이 증가할수록 overfiting에 대한 위험이 크다.  
모델의 과대적합을 방지하는 방법으로 모델에 규제를 가하는 방법을 쓴다.  
다항 회귀의 경우 다항식 차수를 감소시키는 등 자유도를 줄이는 방법이 있다.

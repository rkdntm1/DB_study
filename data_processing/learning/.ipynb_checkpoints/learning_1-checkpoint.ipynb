### 지도 학습 : Supervised Learning
      주어진 입력-결과를 매핑해 주는 함수를 학습
     새로운 데이터x의 출력값 y를 함수 F를 사용하여 예측
     
     
    독립변수(특성 feature)과 목적변수(종속변수, 출력값)간의 연관성을
    표현 또는 예측 즉 f 함수 추정

	F(x)가 이산적(카테고리 범주)일 때 분류(classification)
    F(x) 가 연속적(숫자)일 때 회귀(Regression)
    F(x) 가 x의 확률 p(x)일때 확률추정(Estimation)  

입력-출력의 데이터들로 부터 새로운 입력에 대한 출력을 결정 할 수 있는 패턴 추출  

#### 분류(classification) : 출력이 정해진 부류(class, category) 중 하나로 결정   
    결정트리(decision tree), K-근접이웃 알고리즘(k-nearest neighbor, KNN)    
    서포트 벡터 머신(Support Vector Machine, SVM)  
     랜덤 포레스트(random forest) 등.. 
    
    성능 평가: accuracy(얼마나 정확하게 분류하는가)
               ROC-curve, AUC(area under the curve)
    
#### 회귀분석(regression ): 출력이 연속인 영역의 값 결정
    학습 데이터에 부합되는 출력값이 실수인 함수를 찾는 문제
    
    성능 :최소제곱 : 실제 Y값과 예측 된 Y값의 차이가 최소
          => Sum of Squared Error :SSE




###  비지도학습 :Unsupervised Learning

출력에 대한 정보가 없는 데이터로 부터 필요한 패턴을 추출
   군집(clustering) :  비슷한 것을 묶는 묶는 것  
   비지도 변환(unsupervised transforation) : 데이터를 새롭게 표현하여 사람이나   
      다른 머신러닝 알고리즘이 원래 데이터보다 쉽게 해석할 수 있도록 만드는 알고리즘


#### 용어
 x: 입력(input), 특징(feature), 독립변수(independent value)
       predictor, covariate등으로 부르며 vector 형태   
 y: 결과(response, output), 종속변수(dependent value), target


이산형 : 측정 결과를 셀수 있는 경우  
연속형 : 측정 결과가 무한(infinite) 이 많은 변수

## ROC곡선 and AUC  
   
### ROC( Reciver Operation Characteristic Curve)  
    모든 분류 임계값에서 분류 모델의 성능을 보여주는 그래프  
    FPR (False Positive Rate)이 변할때 TPR(True Positive Rate)이 어떻게 변하는지 나타내는 곡선
    
#### FPR을 x축으로 TPR y축으로 처리
  
#### TPR= TP/ (TP+FN) : 재현율( recall or 민감도 ) 
      실제값 positive(양성) 가 정확이 예측되어야 하는 수준
      => 질병, 오류 등이 있어야 하는 자료를 정확히 예측
      
#### FPR =  FP/(FP+TN)    

### AUC( Area Under the Curve): ROC 곡선 아래 영역
####  AUC가 높다=> 클래스를 구별하는 모델의 성능이 좋음

https://en.wikipedia.org/wiki/Receiver_operating_characteristic
  

 * 0.9 ~ 1 = Excellent(A)   
 * 0.8 ~ 0.9 = Good(B)   
 * 0.7 ~ 0.8 = Fair(C)    
 * 0.6 ~ 0.7 = Poor(D)   
 * 0.5 ~ 0.6 = Fail(F)

부적합(underfitting ) :  학습 데이터를 충분히 학습하지 않은 상태

## 표준화, 정규화

### 표준화 : 각 feature를 평균이 0이고 분산이 1인 가우시안 정규분포를 가진 값으로 변환 
 

### 정규화 : 서로 다른 feature의 크기를 통일하기 위해 크기를 변환해 주는 개념

 #### StandardScaler: 개별 feature를 평균이 0이고 분산이 1인 값으로 변환해 줌  
   => 선형회귀(linear regression), 로지스틱 회귀(logistic regression),     SVN(Support Vector Machine:서포트 벡터머신)   
 
 from sklearn.preprocessing import StandardScaler  
standardScaler = StandardScaler()

이상치가 있다면 평균과 표준편차에 영향을 미쳐서 데이터 확산이 달라지게 된다.
 따라서 이상치가 있는 경우 균형잡힌 척도를 보장할 수 없다.
 

 #### MinMaxScaler : 데이터 값을 0과 1사이 범위의 값을 변환해줌    
from sklearn.preprocessing import MinMaxScaler  
minMaxScaler = MinMaxScaler()  

이상치가 있을 경우 변환된 값이 매우 좁은 범위로 압축될 수 있다.
아웃라이어의 존재에 매우 민감하다

#### <b> $\frac{x-Min(X)}{Max(X)-Min(X)}$ </b> 

 X:데이터 셋
 x:데이터 샘플

 ##### MaxAbsScaler : 최대 절대값과 0이 각각 1,0이 되도록 변환시킴  
  -1~ 1 사이로 재조정되며 큰 이상치에 민감할 수 있다.
  

#### RobustScaler : 중앙값(median)과 IQR(interquartile range)로  
아웃라이너의 영향을 최소화 시켜줌
 표준화 후 동일한 값을 더 넓게 분포시키고 있다.
 
 결론 :  모든 스케일러 처리 전에 아웃라이어 제거가 선행되어야 한다.
 

#### 이상치 : 아웃라이어(outlier)   
통계적 자료분석의 결과를 왜곡시키거나, 자료 분석의 적절성을 위협하는 변수값 또는 사례  
개체의 변수값이 0(평균)으로부터 ±3 표준편차밖에 위치하는 사례 
일반적인 경향에서 벗어나는 사례

```python
from sklearn.datasets import load_iris
import pandas as pd
iris=load_iris()
target=iris.target
data=iris.data
feature=iris.feature_names

iris_df=pd.DataFrame(data=iris.data, columns=feature)

iris_df.head()
```

```python
iris_df.info()
```

```python
iris_df.describe()
```

### standardScaler

```python
from sklearn.preprocessing  import StandardScaler

scaler=StandardScaler()

scaler.fit(iris_df)

iris_scale=scaler.transform(iris_df)

type(iris_scale)  #ndarray

iris_df_scale= pd.DataFrame(iris_scale, columns=iris.feature_names)

iris_df_scale.describe()
```

### MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)
type(iris_scaled)

iris_scale_df=pd.DataFrame(iris_scaled, columns=iris.feature_names)

iris_scale_df.describe()
```

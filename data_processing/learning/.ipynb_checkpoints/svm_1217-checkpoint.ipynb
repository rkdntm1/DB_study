```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
iris= load_iris()
#type(iris)

x=iris['data'][:, [2,3]]
y=iris['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, shuffle = True, random_state = 1000)
```

```python
svc = SVC(kernel='linear', probability=True, decision_function_shape='ovr')
svc.fit(x_train, y_train)
print('train accuracy: ' ,svc.score(x_train,y_train))
print('test accuracy: ' , svc.score(x_test, y_test))
```

```python
import mglearn
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
mglearn.plots.plot_2d_classification(svc, x_train, alpha=0.1)
mglearn.discrete_scatter(x_train[:,0], x_train[:,1], y_train)
plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
print('indicate fof support vectors', svc.support_)
```

```python
svc.support_vectors_
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, recall_score, precision_score
from sklearn.svm.libsvm import predict_proba
```

```python
pred = svc.predict(x_test)

conf_mat = confusion_matrix(y_test,pred)
```

```python
print(conf_mat)
print('accuracy: ' , accuracy_score(y_test, pred ))
print('precision: ', precision_score(y_test, pred,average='macro'))
print('recall : ', recall_score(y_test, pred, average='macro'))
```

```python
from sklearn.model_selection import GridSearchCV

parameters = [{'C':[0.01, 0.1, 0.5, 1, 5, 10, 100, 1000]}]
svm_multi_clf_grid = GridSearchCV(SVC(kernel='linear'),parameters, cv=8, scoring = 'accuracy')
svm_multi_clf_grid.fit(x_train, y_train)

print(svm_multi_clf_grid.best_score_)
print(svm_multi_clf_grid.best_params_)
print(svm_multi_clf_grid.best_index_)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import mglearn
import numpy as np  
from tqdm.notebook import tqdm

iris = load_iris()

x = iris.data[:, [2,3]]
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

optAcc = -999
optG = 0
optX =0
for gamma in tqdm(np.arange(0.1, 10.0, 0.1)):
    for c in np.arange(0.1, 5.0, 0.1):
        model = SVC(kernel='rbf', gamma=gamma, C=c)
        model.fit(x_train, y_train)
        acc = model.score(x_test, y_test)
        
        if acc > optAcc:
            optG = gamma
            optC = c
            optAcc = acc
            
print('optG :', optG)
print('optC :', optC)
print('optAcc :',  optAcc)

svc_model = SVC(kernel='rbf', gamma=optG, C=optC)
svc_model.fit(x_train, y_train)

print()
print('test accruacy',  svc_model.score(x_train, y_train))
print('test accuracy' , svc_model.score(x_test, y_test)) 
```

```python
from sklearn.datasets import load_iris
import pandas as pd

import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
iris=load_iris()
col = ['sepal_length','sepal_width','petal_length','petal_width']
iris_df=pd.DataFrame(iris.data, columns=col)
```

```python
iris_df.head()
```

```python
iris_df.tail()
```

```python
iris_df['target']= iris.target
```

```python
iris_df.head()
```

```python
markers=['^','s','o']
for i, marker in enumerate(markers):
    x_axis_data = iris_df[iris_df['target']==i]['sepal_length']
    y_axis_data = iris_df[iris_df['target']==i]['sepal_width']
    plt.scatter(x_axis_data, y_axis_data, marker=marker, label = iris.target_names[i])
plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
iris_scaled = StandardScaler().fit_transform(iris_df.iloc[:,:-1])
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(iris_scaled)
```

```python
iris_pca = pca.transform(iris_scaled)
print(iris_pca.shape)
```

```python
pca_columns=['pca_component_1', 'pca_component_2']
iris_df_ca = pd.DataFrame(iris_pca, columns=pca_columns)
```

```python
iris_df_ca.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
rfc = RandomForestClassifier(random_state=1000)
scores = cross_val_score(rfc, iris.data, iris.target, scoring ='accuracy', cv = 3)
print('source_cross_val accuracy: ', scores)
print('source_mean accuracy: ', np.mean(scores))
```

```python
iris_df_ca['target']=iris.target
```

```python
markers=['^','s','o']
for i, marker in enumerate(markers):
    x_axis_data = iris_df_ca[iris_df_ca['target']==i]['pca_component_1']
    y_axis_data = iris_df_ca[iris_df_ca['target']==i]['pca_component_2']
    plt.scatter(x_axis_data, y_axis_data, marker=marker, label = iris.target_names[i])
plt.legend()
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
pca_X = iris_df_ca[['pca_component_1', 'pca_component_2']]
scores_pca = cross_val_score(rfc, pca_X, iris.target, scoring='accuracy', cv=3 )
print('PCA trans cross val score:',scores_pca)
print('PCA trans score mean:', np.mean(scores_pca))
```
